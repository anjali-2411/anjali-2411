```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('sales_data.csv',encoding='cp1252')

# Exploratory Data Analysis (EDA)
print(df.head())
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Visualize sales trends over time
plt.figure(figsize=(10, 6))
y=df['PRODUCTLINE']
plt.plot(y)
plt.title('Sales Trend Over Time')
plt.xlabel('DATE')
plt.ylabel('PRODUCTS')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualize distribution of sales by product category
plt.figure(figsize=(10, 6))
x=df['COUNTRY']
sns.boxplot(x)
plt.title('Sales Distribution by Product Category')
plt.show()
```

```python
# Visualize top selling products
top_products = df.groupby('PRODUCTLINE')['SALES'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
top_products.plot(kind='bar')
plt.title('Top Selling Products')
plt.xlabel('Product Name')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
#Analyze correlation between variables
correlation_matrix = df.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')

plt.show()
```
